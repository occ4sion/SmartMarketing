### AssetUser

```python
AU = pd.read_csv('data/AssetUser.csv', index_col='id')
```

```python
size = user.shape[0] * asset.shape[0]
```

```python
index = 0
for a in asset.index:
    for u in user.index:
        AU.loc[index, ['asset', 'user']] = (a, u)
        index += 1
```

```python
AU['conversion'] = AU.apply(
    lambda au: user.at[au.user, 'conversion'] + asset.at[au.asset, 'conversion'], axis=1
)
AU['conversion'] += abs(np.random.normal(size=size)/4)
AU['conversion'] /= AU['conversion'].max()
AU['conversion'] = AU['conversion'].round(3)
```

```python
AU['activity'] = AU.apply(
    lambda au: user.at[au.user, 'activity'] * (asset.at[au.asset, 'cost'] / asset.cost.max()), axis=1
)
AU['activity'] += abs(np.random.normal(size=size)/4)
AU['activity'] /= AU['activity'].max()
AU['activity'] = AU['activity'].round(3)
```

```python
AU['price'] = AU.apply(
    lambda au: user.at[au.user, 'cost'] * asset.at[au.asset, 'cost'], axis=1
)
AU['price'] *= 1e-8
AU['price'] += np.random.normal(20, 50, size=size)
AU['price'] = AU['price'].round()
```

```python
size = AUG.shape[1] - 1
rows = AUG.shape[0]
index = np.arange(0, rows)
```

```python
def get_price(aug):
    idx = aug.T.iloc[:,].copy().values[:-1]
    idx = idx.T * np.arange(0, size)
    idx = idx[idx != 0]
    return AU.loc[idx, 'price'].sum()
```

```python
AUG.loc[index, '0_AU'] = 0
AUG.loc[index, AUG.columns[:-1]] = np.random.binomial(
    n=1, p=.001+np.random.exponential(scale=1e-3, size=rows), size=(size, rows)
).T
idx = AUG.T.iloc[:, index].copy().values[:-1]
idx = idx.T * np.arange(0, size)
idx = idx[idx != 0]
AUG.loc[index, 'price'] = AUG.apply(get_price, axis=1)
```

### Campaign

```python
campaign = pd.read_csv('data/campaign.csv', index_col='id')
```

```python
size = product.shape[0]
```

```python
campaign['product'] = product.index.values
```

```python
prob = abs(np.random.normal(scale=2, size=AUG.index.size))
campaign['AUG'] = np.random.choice(AUG.index, size=size, p=prob/prob.sum())
```

```python
campaign['rounds'] = 1 + np.random.binomial(3, p=np.repeat([.6, .2, .1, .1], size/4), size=size)
```

```python
addition = ['design', 'photos', 'movie', 'site', 'accounts']
campaign[addition] = np.random.binomial(1, p=[.15, .25, .05, .25, .3], size=(size, 5)) * np.array([4000, 1500, 4000, 3000, 800])
```

```python
campaign['budget'] = AUG.loc[campaign.AUG, 'price'].values + campaign[addition].sum(1)
```

```python
campaign['budget'] = np.around(campaign['budget'], -3)
```

```python
campaign['start'] = np.random.choice(pd.date_range('01-01-2022', '25-11-2022', freq='D'), size=size)
```

```python
def scale(vector):
    return (vector - vector.min()) / vector.max()
def bins(vector, nbins=2):
    _, edges = np.histogram(vector, nbins)
    return (vector >= edges[:, np.newaxis]).sum(0)
```

$Revenue = Price * \#Users * Conversion(user, product) * volume$

```python
AUG_ = AUG.loc[campaign.AUG, :].drop('price', axis=1)

price = np.around(0
    + product.loc[campaign['product'], 'price'].values 
    + np.random.normal(100, 30, size=size)
, -1)

users = scale(1
    * 1 * company.loc[product.loc[campaign['product'], 'company'], 'popularity'].values
    * 2 * product.loc[campaign['product'], ['novelty', 'popularity', 'conversion']].values.sum(1)
    * 3 * (AUG_ * user.loc[AU.loc[:, 'user'], 'activity'].values.T).sum(1).values
    * 1 * (AUG_ * AU.loc[:, 'conversion'].values.T).sum(1).values
)

conversion = scale(1
    * 2 * (product.loc[campaign['product'], ['conversion', 'popularity']].values * np.array([1, .75])).sum(1)
    * 1 * (AUG_ * AU.loc[:, 'conversion'].values.T).sum(1).values
    * .6 * (AUG_ * user.loc[AU.loc[:, 'user'], 'conversion'].values.T).sum(1).values
)
volume = product.loc[campaign['product'], 'volume'].values

revenue = 25000 + (.5 * price) * (.2 * campaign.rounds.values + 2 * users) * (3 * conversion) * (.1 * volume)
revenue.shape
```

```python
campaign['revenue'] = np.around(revenue, -3)
#campaign.loc[campaign['revenue'] > 1.5 * campaign['budget'], 'revenue'] - campaign['budget'] * ((campaign['revenue'] / campaign['budget']) - 1)
```
